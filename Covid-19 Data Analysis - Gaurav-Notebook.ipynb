## Date wise trend of positive cases on ICMR Testing data set

```python
fig, ax = plt.subplots(figsize = (12,6))    
fig = sns.barplot(x="day", y="totalPositiveCases", data = df_test, 
                  estimator = sum, ci = None, ax=ax)

x_dates = df_test['day'].dt.strftime('%Y-%m-%d').sort_values().unique()
ax.set_xticklabels(labels=x_dates, rotation=45, ha='right')
```





## Correlation testing - LM plot - United States Data set

```python
sns.lmplot(x='hospitalizedCumulative', y='onVentilatorCumulative', data=df_US, ci=None , line_kws={"color":"g","alpha":0.7,"lw":5})
core = df_US[['hospitalizedCumulative','onVentilatorCumulative']].corr()
core
```





## Joint Grid Plot - US Data Set

```python
sns.set_theme(context='notebook', style='darkgrid', palette='deep')
k = sns.JointGrid(data=df_US, x="positive", y="hospitalizedCurrently", height=12)
k.plot_joint(sns.scatterplot, s=200, alpha=1 , color = 'purple')
k.plot_marginals(sns.histplot, kde=True , color = 'red')
```

```python
sns.jointplot(data=df_US, x="positive", y="recovered", height=10, kind="kde" , color = 'green')
```

## SEABORN Heat map and Correlation - US data set

```python
corr = df_US[['positive', 'negative', 'pending', 'hospitalizedCurrently', 'hospitalizedCumulative', 'inIcuCurrently', 'inIcuCumulative', 'onVentilatorCurrently', 'onVentilatorCumulative', 'recovered']].corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
plt.figure(figsize=(14, 14))
heatmap = sns.heatmap(corr,cmap='viridis', xticklabels=corr.columns,yticklabels=corr.columns)
plt.show()
```
